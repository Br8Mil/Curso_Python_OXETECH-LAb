```python
import PySimpleGUI as sg

interface = [
[sg.Text('User')], #Função para criar uma string.
[sg.Input(key='user')], #É a função que permite ao usúario digitar seu login, e me permite manipular seus dados.
[sg.Text('Password')],
[sg.Input(key='pass')],
[sg.Button('Login')], #Função para criar botões na interface.
[sg.Text('', key= 'message')],
]

window = sg.Window('Login', layout=interface)
#sg.WIndows e uma uma função que precisa de dois paramêtros:
#1° O nome da janela.
#2° É o meu layout.

while True:
    event, values = window.read()
    #Evento é todo o local em que posso clicar na interface.
    #Valores serão todos os cantos em que posso manipular.
    if event == sg.WIN_CLOSED: #sg.WIN_CLOSED e igual a clicar no botão de fechar.
        break
    elif event == 'Login':
        user_login = values['user']
        pass_login = values['pass']
        user_correct = 'Rodrigo'
        pass_correct = '12345'
        if user_login == user_correct and pass_login == pass_correct:
            window['message'].update('Successful Login')
        else:
            window['message'].update('username or password is invalid')
```
